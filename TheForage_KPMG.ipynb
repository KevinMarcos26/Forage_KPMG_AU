**MODELO 2**

```python
from sklearn.metrics import classification_report
```

```python
import xgboost as xgb
```

```python
# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=26)

# Calcular el cociente entre el número de muestras negativas y el número de muestras positivas en el conjunto de entrenamiento
ratio = len(y_train[y_train==0])/len(y_train[y_train==1])

# Definir el parámetro scale_pos_weight de XGBClassifier
model = xgb.XGBClassifier(n_estimators=500, learning_rate=0.05, n_jobs=4,scale_pos_weight=ratio,random_state=26)
```

```python
# Entrenar el modelo con los datos de entrenamiento ajustados
model.fit(x_Train, y_Train, 
             early_stopping_rounds=9, 
             eval_set=[(x_Test, y_Test)], 
             verbose=False)

# Realizar la predicción en el conjunto de prueba
y_pred = model.predict(X_test)

# Imprimir el reporte de clasificación
print(classification_report(y_test, y_pred))
```

```python
#test model
accuracy_clf = accuracy_score(y_Test, y_pred)
auc_roc1 = roc_auc_score(y_Test, y_pred)
print("Accuracy score of the model is: {}".format(accuracy_clf))
print("roc_auc_score of the model1 is: {}".format(auc_roc1))
```
